```python
from selenium import webdriver
import requests
import time
from bs4 import BeautifulSoup

# Configura o Selenium com o navegador Chrome
driver = webdriver.Chrome()

# Acessa a página inicial para obter os cookies
driver.get('https://legislacao.presidencia.gov.br')

# Espera alguns segundos para garantir que a página seja carregada completamente
driver.implicitly_wait(60)

# Atualiza a página para garantir que os cookies mais recentes sejam obtidos
driver.refresh()

# Espera mais alguns segundos para garantir que a página tenha sido atualizada
time.sleep(15)  # Ajuste o tempo se necessário

# Obtém os cookies do navegador
selenium_cookies = driver.get_cookies()
print(selenium_cookies)

# Fecha o navegador
driver.quit()

# Inicia uma sessão do requests
session = requests.Session()

# Adiciona os cookies obtidos pelo Selenium na sessão do requests
for cookie in selenium_cookies:
    session.cookies.set(cookie['name'], cookie['value'])

# Formata os cookies como uma string para o header 'Cookie'
cookie_header = "; ".join([f"{cookie['name']}={cookie['value']}" for cookie in selenium_cookies])
print(cookie_header)

# Define os headers para a requisição POST, incluindo o header de cookies formatado
headers = {
    "Accept": "*/*",
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Accept-Language": "pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7",
    "Cache-Control": "no-cache",
    "Connection": "keep-alive",
    "Content-Type": "application/x-www-form-urlencoded; charset=UTF-8",
    "Cookie": f"{cookie_header}",
    "Host": "legislacao.presidencia.gov.br",
    "Origin": "https://legislacao.presidencia.gov.br",
    "Pragma": "no-cache",
    "Referer": "https://legislacao.presidencia.gov.br/",
    "Sec-CH-UA": '"Not)A;Brand";v="99", "Google Chrome";v="127", "Chromium";v="127"',
    "Sec-CH-UA-Mobile": "?1",
    "Sec-CH-UA-Platform": '"Android"',
    "Sec-Fetch-Dest": "empty",
    "Sec-Fetch-Mode": "cors",
    "Sec-Fetch-Site": "same-origin",
    "User-Agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Mobile Safari/537.36",
    "X-Requested-With": "XMLHttpRequest"
}

# Define o payload para a requisição POST
payload = {
    "pagina": 0,
    "posicao": 0,
    "termo": "saúde",
    "ano_ato": 2018,
    "ordenacao": "maior_data"
}

# URL para a requisição POST
url = "https://legislacao.presidencia.gov.br/pesquisa/ajax/resultado_pesquisa_legislacao.php"

# Faz uma requisição POST usando a sessão com os cookies já armazenados
response = session.post(url, data=payload, headers=headers)

# Verifica a resposta da requisição
print("Resposta da requisição:", response.text)
```

```python
import re


html_response = response.text

soup = BeautifulSoup(html_response, 'html.parser')
#print(soup)

# Usa uma expressão regular para encontrar todos os valores de href
href_pattern = re.compile(r'href=["\'](https?://[^"\']+)["\']')

# Encontra todos os hrefs no HTML
links = href_pattern.findall(html_response)


list_links = []
# Exibe os links encontrados
for link in links:
    list_links.append(link)
```

```python
list_links = list_links[0:1]
```

```python
import requests
from bs4 import BeautifulSoup
import json

# Headers para a requisição
headers = {
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Accept-Language": "pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7",
    "Cache-Control": "no-cache",
    "Connection": "keep-alive",
    "Cookie":  f"{cookie_header}",
    "Host": "www.planalto.gov.br",
    "Pragma": "no-cache",
    "Sec-CH-UA": '"Not)A;Brand";v="99", "Google Chrome";v="127", "Chromium";v="127"',
    "Sec-CH-UA-Mobile": "?1",
    "Sec-CH-UA-Platform": '"Android"',
    "Sec-Fetch-Dest": "document",
    "Sec-Fetch-Mode": "navigate",
    "Sec-Fetch-Site": "none",
    "Sec-Fetch-User": "?1",
    "Upgrade-Insecure-Requests": "1",
    "User-Agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Mobile Safari/537.36"
}

# Lista de links para acessar
links = list_links

# Dicionário para armazenar o conteúdo dos links
content_dict = {}

for link in links:
    # Faz a requisição para o link
    response = requests.get(link, headers=headers)
    
    # Verifica se a requisição foi bem-sucedida
    if response.status_code == 200:
        # Cria o BeautifulSoup para extrair o texto
        soup = BeautifulSoup(response.content, 'html.parser')
        text = soup.get_text(separator='\n', strip=True)  # Extrai o texto
        content_dict[link] = text
    else:
        content_dict[link] = f"Erro ao acessar o link: {response.status_code}"

# Salva o conteúdo em um arquivo JSON
with open('content.json', 'w', encoding='utf-8') as f:
    json.dump(content_dict, f, ensure_ascii=False, indent=4)

print("Conteúdo salvo em content.json")
```
